```python
import pandas as pd
import nltk
import json
```

```python
def open_json(file_path: str):
    with open(file_path, 'r') as file:
        return json.load(file)
```

```python
image_data = open_json('scraping/recipes/recipes_with_images.json')
image_data = pd.DataFrame(image_data)
image_data.head()
```

```python
# removing rows where there are no directions
image_data = image_data[image_data['directions'].apply(lambda x: x != [])]

# removing unnecessary columns
image_data = image_data.drop(columns=['prep_time','refrigerate_time', 'stand_time', 'cool_time', 'marinate_time', 'additional_time', 'rest_time', 'chill_time',
       'freeze_time', 'rise_time', 'active_time', 'steep_time',
       'slow_cook_time', 'soak_time', 'other_time_time', 'hands_on_time',
       'release_pressure_time', 'bake_time'])
```

```python
image_data.head(5)
```

```python
recipes = pd.read_csv('data/recipe_data.csv')
recipes.head()
```

```python
merged = recipes.merge(image_data[['title', 'image_filename']], on='title', how='left')
```
