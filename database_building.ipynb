```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('recipes.db')
```

```python
curs = conn.cursor()
```

```python
recipes = pd.read_csv('data/recipe_data.csv')
```

```python
recipes.to_sql('recipes', 
            conn,
            index=False)
```

```python
# Select all of the data
pd.read_sql("""
SELECT * 
FROM recipes
LIMIT 10
;""", conn)
```

```python
conn.close()
```
